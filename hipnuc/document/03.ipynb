Trong không gian 3D, chuyển động quay quanh một trục cố định được gọi là chuyển động quay sơ cấp Givens, bất kỳ chuyển động quay nào trong không gian đều có thể được tổng hợp bởi ba phép quay cơ bản này.

```python
import numpy as np

cos = np.cos
sin = np.sin
pi = np.pi
```

```python
def ch_rotx(theta):
    return np.array([[1, 0, 0], 
                     [0, cos(theta), -sin(theta)], 
                     [0, sin(theta), cos(theta)]])

def ch_roty(theta):
    return np.array([[cos(theta), 0, sin(theta)], 
                     [0, 1, 0], 
                     [-sin(theta), 0, cos(theta)]])

def ch_rotz(theta):
    return np.array([[cos(theta), -sin(theta), 0], 
                     [sin(theta), cos(theta), 0], 
                     [0, 0, 1]])
```

```python
i = 30
i_rad = np.deg2rad(i)
Cb2nX = ch_rotx(i_rad)
print("Rotate around the X axis %.1f ° the rotation matrix of Cb2nX:\n" %(i), Cb2nX)

i = 40
i_rad = np.deg2rad(i)
Cb2nY = ch_roty(i_rad)
print("Rotate around the Y axis %.1f ° the rotation matrix of Cb2nY:\n" %(i), Cb2nY)

i = 50
i_rad = np.deg2rad(i)
Cb2nZ = ch_rotz(i_rad)
print("Rotate around the Z axis %.1f ° the rotation matrix of Cb2nZ:\n" %(i), Cb2nZ)
```

Chuyển một điểm tọa độ A từ hệ b sang hệ n thông qua các ma trận sau:

```python
Ab = np.array([0, 0, 1])

print("point A belongs to the coordinate b system:\n", Ab)

An = Cb2nX.dot(Ab)
print("The point A under the b system is rotated to the n system by Cb2nX:\n", An)

An = Cb2nY.dot(Ab)
print("The point A under the b system is rotated to the n system by Cb2nY: \n", An)

An = Cb2nZ.dot(Ab)
print("The point A under the b system is rotated to the n system by Cb2nZ: \n", An)
```

Nâng cao:

- Hãy thay đổi góc quay của các ma trận thành -90 °, rồi chuyển điểm A [0,0,1] trong hệ thống b qua ba ma trận Cb2n được quay -90 ° để nhận kết quả.
- Sau đó thử di chuyển lần nữa, cầm một cây bút và xoay nó theo cách thủ công trong trí tưởng tượng của bạn và so sánh kết quả. 

Thông qua bài này bạn sẽ hiểu sâu hơn về phép biến đổi ma trận cơ bản 3D.

```python
Ab = np.array([1, 0, 0])

print("point A belongs to the coordinate b system:\n", Ab)

An = ch_roty(np.deg2rad(-90)).dot(Ab)
print("The point A under the b system is rotated to the n system by Cb2nY: \n", An)

An = ch_rotx(np.deg2rad(-90)).dot(An)
print("The point A under the b system is rotated to the n system by Cb2nX:\n", An)

An = ch_rotz(np.deg2rad(-90)).dot(An)
print("The point A under the b system is rotated to the n system by Cb2nZ: \n", An)

# ch_rotz(-90).dot(ch_rotx(-90)).dot(ch_roty(-90)).dot(Ab)
```
