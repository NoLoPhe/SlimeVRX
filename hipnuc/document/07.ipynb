```python
import numpy as np
from numpy.linalg import norm   #norm(V,2)
```

```python
def ch_qconj(qin):
    return np.concatenate([[qin[0]], -qin[1:]])

def ch_qnormlz(q):
    q = q / norm(q,2)
    if q[0] < 0:
        q[0] = -q[0]
        q[1] = -q[1]
        q[2] = -q[2]
        q[3] = -q[3]
    return q

def ch_qmul(q1, q2):
    q = np.array([ q1[0] * q2[0] - q1[1] * q2[1] - q1[2] * q2[2] - q1[3] * q2[3],
                   q1[0] * q2[1] + q1[1] * q2[0] + q1[2] * q2[3] - q1[3] * q2[2],
                   q1[0] * q2[2] + q1[2] * q2[0] + q1[3] * q2[1] - q1[1] * q2[3],
                   q1[0] * q2[3] + q1[3] * q2[0] + q1[1] * q2[2] - q1[2] * q2[1] ], dtype=np.float64)
    return q

def ch_q2m(Qb2n):
    q11 = Qb2n[0]*Qb2n[0]; q12 = Qb2n[0]*Qb2n[1]; q13 = Qb2n[0]*Qb2n[2]; q14 = Qb2n[0]*Qb2n[3]
    q22 = Qb2n[1]*Qb2n[1]; q23 = Qb2n[1]*Qb2n[2]; q24 = Qb2n[1]*Qb2n[3]
    q33 = Qb2n[2]*Qb2n[2]; q34 = Qb2n[2]*Qb2n[3]
    q44 = Qb2n[3]*Qb2n[3]
    Cb2n = np.array([[ q11+q22-q33-q44,  2*(q23-q14),     2*(q24+q13)],
                     [2*(q23+q14),      q11-q22+q33-q44, 2*(q34-q12)],
                     [2*(q24-q13),      2*(q34+q12),     q11-q22-q33+q44 ]], dtype=np.float64)
    return Cb2n

def ch_m2q(Cb2n):
    C11 = Cb2n[0,0]; C12 = Cb2n[0,1]; C13 = Cb2n[0,2]
    C21 = Cb2n[1,0]; C22 = Cb2n[1,1]; C23 = Cb2n[1,2]
    C31 = Cb2n[2,0]; C32 = Cb2n[2,1]; C33 = Cb2n[2,2]

    if C11 >= C22+C33:
        q1 = 0.5*np.sqrt(1+C11-C22-C33)
        q0 = (C32-C23)/(4*q1); q2 = (C12+C21)/(4*q1); q3 = (C13+C31)/(4*q1)
    elif C22>= C11+C33:
        q2 = 0.5*np.sqrt(1-C11+C22-C33)
        q0 = (C13-C31)/(4*q2); q1 = (C12+C21)/(4*q2); q3 = (C23+C32)/(4*q2)
    elif C33>=C11+C22:
        q3 = 0.5*np.sqrt(1-C11-C22+C33)
        q0 = (C21-C12)/(4*q3); q1 = (C13+C31)/(4*q3); q2 = (C23+C32)/(4*q3)
    else:
        q0 = 0.5*np.sqrt(1+C11+C22+C33)
        q1 = (C32-C23)/(4*q0); q2 = (C13-C31)/(4*q0); q3 = (C21-C12)/(4*q0)
    Qb2n = np.array([q0, q1, q2, q3], dtype=np.float64)
    return Qb2n
```

```python
Qb2n = np.array([0.981, 0.010, -0.191, 0.011])
Cb2n = ch_q2m(Qb2n)
print("Quaternion Qb2n to Attitude Array Cb2n:\n", Cb2n)

Qb2n = ch_m2q(Cb2n)
print("Attitude array Cb2n to Quaternion Qb2n:\n", Qb2n)
```

Ý nghĩa của toán tử (nhân) về cơ bản là để thực hiện phép nhân Quaternion trước, vì vậy xử dựng Quaternion để xoay, và sau đó trích xuất phần vectơ ảo của nó.


```python
def ch_qmulv(q, vin):
    qi = np.concatenate([[0], vin])
    qo = ch_qmul(ch_qmul(q,qi),ch_qconj(q))
    vout = qo[1:]
    return vout
```

```python
# Test 1: b2n conversion
Qb2n = np.array([0.919, 0.262,  -0.09, -0.282])

# accReading: Giá trị đọc được khi gia tốc kế ở trạng thái nghỉ.
# tương đương với giá trị gia tốc tĩnh trong hệ thống b (vectơ ba chiều trong hệ thống b)
accReading = np.array([0.018, 0.531,  0.843])

Cb2n = ch_q2m(Qb2n)
Gn = Cb2n.dot(accReading)

# Gia tốc của hệ b được chuyển đổi sang hệ n thông qua phép Cb2n
print("Eg1: Acceleration of the b system is transferred to the n Cb2n, the result: %.3f %.3f %.3f \n" %(Gn[0], Gn[1], Gn[2]))
# Kết quả gần bằng 0 0 1

Gn = ch_qmulv(Qb2n, accReading)
# Dùng Quaternion
print("Eg2: Use Quaternion to accReading, get the same result: %.3f %.3f %.3f \n" %(Gn[0], Gn[1], Gn[2]))

# Test 2: n2b conversion
Gn = np.array([0, 0, 1])
# Gn: Ở trạng thái nghỉ, giá trị gia tốc trong hệ N (bằng trọng lực, ngược chiều)
Qn2b = ch_qconj(Qb2n)
Cn2b = ch_q2m(Qn2b)

accB = Cn2b.dot(Gn)

# Gia tốc của hệ n được chuyển đổi sang hệ b thông qua phép Cn2b
print("Eg1: transferred to the b Cn2b, the result: %.3f %.3f %.3f" %(accB[0], accB[1], accB[2]))
print("accReading: %.3f %.3f %.3f\n" %(accReading[0], accReading[1], accReading[2]))

accB = ch_qmulv(Qn2b, Gn)
# Dùng Quaternion rotation function
print("Eg2: Using the Quaternion rotation function, get the same result: %.3f %.3f %.3f \n" %(accB[0], accB[1], accB[2]))
```

Thử nghiệm này có thể nâng cao một chút, khái niệm gia tốc được giới thiệu. Đọc giá trị gia tốc của cảm biến ở trạng thái tĩnh là vectơ tương đương hệ b và trọng lực lý thuyêt tương đương hệ n là [0 0 1]. Chỉ cần nghĩ chúng như là vectơ 3D thông thường.
