### Gia tốc kế, Con quay hồi chuyển, Từ kế

Gia tốc kế là một cảm biến quán tính được sử dụng phổ biến để đo gia tốc tuyến tính + gia tốc. Cái gọi là tỷ lệ.

<p align="left"><img width="500", src="https://pic1.zhimg.com/80/v2-5fcfc45c8da2fb4bc8a49aed53b3a234_720w.jpg"></p>

Gia tốc kế đo gia tốc cảm nhận theo ba trục, vì vậy khi đặt gia tốc kế nằm ngang và đứng yên trên trái đất, đọc là 0,0,1G. Khi gia tốc kế nhận được gia tốc 1G theo chiều dương của trục X, đọc là 1G,0,1G

Có một chút khó khăn ở đây. Trọng lực hướng xuống theo phương thẳng đứng, nhưng mặt trước được đặt theo phương ngang, và đọc là 0,0,1G, tại sao không phải 0,0, -1G. Điều này là do gia tốc kế cảm nhận được lực từ bên ngoài (đặt gia tốc kế như mô hình lò xo). Thay vì một trường hấp dẫn.

### Con quay hồi chuyển

Con quay hồi chuyển đo vận tốc góc ba trục, nói thẳng ra là nó đo chuyển động quay quanh mỗi trục, nó còn được gọi là cảm biến vận tốc góc và cũng là một thiết bị quán tính.

<p align="left"><img width="500", src="https://pic3.zhimg.com/80/v2-0beea6fb86cdff02cfdd90dbceeb81ee_720w.jpg"></p>

Gia tốc kế + con quay hồi chuyển được gọi là IMU (Đơn vị đo lường quán tính), được gọi là cảm biến 6 trục

### Từ trường của trái đất

Sự phân bố hình học của từ trường trái đất rất không đều. Nói chung, các đường sức từ bắt đầu từ cực nam và quay trở lại cực bắc của từ trường:

<p align="left"><img width="500", src="https://pic3.zhimg.com/80/v2-3f833327ef2ccce3801ba06796a99f1e_720w.jpg"></p>

Chú ý đến chiều của các đường sức từ: từ cực nam đến cực bắc của đường sức từ.

Đối với bán cầu bắc, hướng của địa từ nghiêng xuống và đối với bán cầu nam, hướng của địa từ nghiêng lên trên.

<p align="left"><img width="400", src="https://pic1.zhimg.com/80/v2-c41b17a750127737ed9f909c42636fb4_720w.jpg"></p>

Một trường địa từ điển hình ở bán cầu bắc

Dòng của Cường độ là giá trị đọc của cảm biến từ trường 3 trục, là vectơ không gian ba chiều. Môđun của Cường độ là cường độ từ trường:

<p align="left"><img width="200", src="https://www.zhihu.com/equation?tex=%5Csqrt%7BMx%5E%7B2%7D+%2B+My%5E%7B2%7D%2B+Mz%5E%7B2%7D%7D"></p>

- Declination: Góc giữa từ trường Trái đất và hướng bắc địa lý.
- Inclination: Góc giữa từ trường Trái đất và phương nằm ngang.
- Chuyển đổi đơn vị: Đơn vị phổ biến: MicroTesla (uT), MilliGauss (mGauss)
- 1 uT = 10 mGauss, phạm vi của trường địa từ: 250 - 650 mGauss hoặc 25 - 64 uT
