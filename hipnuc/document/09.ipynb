thời gian. Tuy nhiên trong thực tế không tồn tại trục quay cố định tuyệt đối, tăng tần số tích phân (giảm khoảng tích phân) là một cách, ngoài ra còn có nhiều thuật toán mẫu con để thu được một mức độ bù. Trên thực tế, tần số tích hợp của con quay hồi chuyển giá rẻ MEMS về cơ bản là 100-1000Hz. Cũng không cần áp dụng thuật toán nhiều mẫu để bù (nhiễu của chính con quay hồi chuyển, sai số sai lệch 0 lớn hơn nhiều so với lỗi thuật toán).

```python
import numpy as np
sin = np.sin
cos = np.cos
atan2 = np.arctan2
asin = np.arcsin
```

```python
def ch_askew(v):
    m = np.array([[0,     -v[2],   v[1]],
                  [v[2],   0,     -v[0]],
                  [-v[1],   v[0],   0]], dtype=np.float64)
    return m

def ch_rv2m(rv):
    nm2 = rv.dot(rv)
    if nm2 < 1.e-8:     #không hiểu
        a = 1 - nm2 * (1 / 6 - nm2 / 120)
        b = 0.5 - nm2 * (1 / 24 - nm2 / 720)
    else:
        nm = np.sqrt(nm2)
        a = sin(nm) / nm
        b = (1 - cos(nm)) / nm2
    VX = ch_askew(rv)
    m = np.identity(3) + a * VX + b * VX.dot(VX)
    return m

def ch_mnormlz(Cb2n):
    for i in range(4):
        # Cnb = 0.5 * (Cnb + (Cnb')^-1)                       # Algorithm 1 
        Cb2n = 1.5*Cb2n - 0.5*(Cb2n.dot(Cb2n.T)).dot(Cb2n)    # Algorithm 2 
    return Cb2n

def ch_m2eul_321(Cb2n):
    beta = atan2(Cb2n[2,1], Cb2n[2,2])
    alpha = asin(-Cb2n[2,0])
    gamma = atan2(Cb2n[1,0],Cb2n[0,0])
    return beta, alpha, gamma

def ch_m2eul_312(Cb2n):
    beta = asin(Cb2n[2,1])
    alpha = atan2(-Cb2n[2,0], Cb2n[2,2])
    gamma = atan2(-Cb2n[0,1], Cb2n[1,1])
    return beta, alpha, gamma
```

```python
w = np.array([2, -3, 5])
dt = 0.01   # Thời gian cập nhật thái độ: 0,01 giây = 100Hz
dur = 100   # Thời lượng duration

N = int(dur / dt)   # điểm points
eul = np.zeros((N, 3), dtype=np.float64)
# vận tốc góc hệ b (đầu ra con quay hồi chuyển)
print("Angular velocity (gyro output is): %.3fdeg/s %.3fdeg/s %.3fdeg/s\n" %(w[0], w[1], w[2]))

Cb2n = np.identity(3)
theta = np.deg2rad(w)*dt    # Theo trục quay cố định: Vectơ quay tương đương = gia số góc = vận tốc góc * dt

for i in range(N):
    C_m2m_1 = ch_rv2m(theta)
    Cb2n = Cb2n.dot(C_m2m_1)
    Cb2n = ch_mnormlz(Cb2n)
    # a, b, c = ch_m2eul_312(Cb2n)
    # print("Quaternion to Euler angles: %.3f %.3f %.3f " % (np.rad2deg(a), np.rad2deg(b), np.rad2deg(c)))
    eul[i,:] = np.rad2deg(ch_m2eul_312(Cb2n))
```

```python
import matplotlib.pyplot as plt

fig = plt.figure()
plt.plot(eul)
ax = plt.subplot()
ax.legend(["PITCH(deg)", "ROLL(deg)", "YAW(deg)"])
```

Đây là thuật toán cập nhật vectơ quay tương đương theo sau (trong ví dụ trên, vectơ quay tương đương chỉ đơn giản là gần đúng như vận tốc góc * dt)
- ch_rv2m đã được đề cập trong chương trước.

### Phương trình và giải pháp vi phân Quaternion

Công thức cập nhật Attitude của Quaternion không được suy luận lại, và bạn có thể tham khảo chi tiết trong sách giáo khoa của giáo viên Yan. Như đã đề cập trước đó, Quaternion là "phiên bản nén" của mảng Attitude. Tất cả các thuộc tính đều giống với mảng Attitude. Công thức được đưa ra trực tiếp ở đây:

<p align="left"><img width="300", src="https://pic4.zhimg.com/80/v2-3ae1002a2226796305da86d0ca6eba6f_720w.png"></p>

Công thức cập nhật ma trận Attitude Quaternion

Các tên tượng trưng như số gia góc vẫn giống như trước đây. So với công thức cập nhật ma trận Attitude, không khó để nhận thấy rằng cả hai rất giống nhau về hình thức, nhưng tính toán và dung lượng lưu trữ của thuật toán Quaternion sẽ nhỏ hơn một chút, bằng khoảng một nửa ma trận Attitude.

```python
def ch_rv2q(rv):
    nm2 = rv.dot(rv)    # nm2 = rv'*rv 
    if nm2 < 1.e-8:
        q0 = 1-nm2*(1/8-nm2/384)
        s = 1/2-nm2*(1/48-nm2/3840)
    else:
        nm = np.sqrt(nm2)
        q0 = cos(nm/2)
        s = sin(nm/2)/nm
    q = np.concatenate([[q0],s*rv])
    return q

def ch_qmul(q1, q2):
    q = np.array([ q1[0] * q2[0] - q1[1] * q2[1] - q1[2] * q2[2] - q1[3] * q2[3],
                   q1[0] * q2[1] + q1[1] * q2[0] + q1[2] * q2[3] - q1[3] * q2[2],
                   q1[0] * q2[2] + q1[2] * q2[0] + q1[3] * q2[1] - q1[1] * q2[3],
                   q1[0] * q2[3] + q1[3] * q2[0] + q1[1] * q2[2] - q1[2] * q2[1] ], dtype=np.float64)
    return q

from numpy.linalg import norm
def ch_qnormlz(q):
    # Chuẩn hóa / Quaternion normalization
    q = q / norm(q,2)
    if q[0] < 0:
        q[0] = -q[0]
        q[1] = -q[1]
        q[2] = -q[2]
        q[3] = -q[3]
    return q

def ch_q2eul_312(Qb2n):
    q0 = Qb2n[0]
    q1 = Qb2n[1]
    q2 = Qb2n[2]
    q3 = Qb2n[3]

    roll =  -atan2( 2*( q1*q3 - q0*q2 ) , q0*q0 - q1*q1 - q2*q2 + q3*q3)
    pitch = asin( 2*(q0*q1 + q2*q3) )
    yaw = -atan2(2*( q1*q2 - q0*q3 ), q0*q0 - q1*q1 + q2*q2 - q3*q3)
    return pitch, roll, yaw
```

```python
w = np.array([2, -3, 5])
dt = 0.01   # Thời gian cập nhật thái độ: 0,01 giây = 100Hz
dur = 100   # Thời lượng duration

N = int(dur / dt)   # điểm points
eul_ = np.zeros((N, 3), dtype=np.float64)
# vận tốc góc hệ b (đầu ra con quay hồi chuyển)
print("Angular velocity (gyro output is): %.3fdeg/s %.3fdeg/s %.3fdeg/s\n" %(w[0], w[1], w[2]))

Qb2n = np.array([1, 0, 0, 0])
theta = np.deg2rad(w)*dt    # Theo trục quay cố định: Vectơ quay tương đương = gia số góc = vận tốc góc * dt

for i in range(N):
    Q_m2m_1 = ch_rv2q(theta)
    Qb2n  = ch_qmul(Qb2n, Q_m2m_1)
    
    # Quaternion unitization
    Qb2n = ch_qnormlz(Qb2n)
    eul_[i,:] = np.rad2deg(ch_q2eul_312(Qb2n))

fig = plt.figure()
plt.plot(eul_)
ax = plt.subplot()
ax.legend(["PITCH(deg)", "ROLL(deg)", "YAW(deg)"])
```

Tóm tắt
- Phương trình vi phân của ma trận thái độ và Quaternion, trạng thái rời rạc, và công thức đệ quy theo giả thiết quay trục cố định là những công thức quan trọng nhất của tích phân con quay hồi chuyển, những công thức này phải được ghi nhớ.
- Quaternion thường được sử dụng trong các dự án thực tế, tính toán nhanh và tính theo đơn vị tốt. Thoải mái hơn mảng Attitude.
- Trong các ứng dụng thực tế, nó có thể được coi là chuyển động quay theo trục cố định trong một khoảng thời gian ngắn.
- Bản thân công thức trên sử dụng RV, nhưng giá trị này được tính gần đúng bằng chuyển động quay của trục cố định và vận tốc góc * dt được tính gần đúng là RV.
- ch_mnormlz và ch_qnormlz là các hàm chuẩn hóa. Do lỗi cắt bớt độ chính xác do tích hợp lâu dài gây ra, mảng thái độ sẽ mất dần tính xác định dương và hàm Quaternion sẽ mất dần thuộc tính của đơn vị Quaternion (Môđun không phải là một, chỉ có đơn vị Quaternion có môđun là 1 mới có thể đại diện cho Attitude).
